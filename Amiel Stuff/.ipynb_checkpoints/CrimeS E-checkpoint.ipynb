```python
df = pd.read_csv('data/hpq_mem.csv', dtype = dtypes)
df.head()
```

### Get the total reported crimes in the dataset

```python
dfAllCrimes = df.loc[df["mcrimeind"] == 1]
```

```python
dfAllCrimes.head()
```

```python
print("There are " + str(len(dfAllCrimes)) + " reported crimes")
print("There are " + str(len(dfAllCrimes.loc[dfAllCrimes["brgy_resid_code"].isnull() & dfAllCrimes["brgy_resid_txt"].isnull()])) + " untracable crimes")
```

### Get the rows where "brgy_resid_txt" isn't null

```python
dfBarangayText = df[['brgy_resid_txt', 'mcrimeind',
       'mtheftind', 'mrapeind', 'minjurind', 'mcarnapind', 'mcattrustlind',
       'mocrimind', 'mocrim', 'mtheftloc', 'mrapeloc', 'minjurloc',
       'mcarnaploc', 'mcattrustlloc', 'mocrimloc']]
dfBarangayText = dfBarangayText[dfBarangayText.brgy_resid_txt.notnull()]
dfBarangayText.head()
```

### Get the rows where mcrimeind = 1 because we only want to see the crimes that happened

```python
dfCrimesText = dfBarangayText[dfBarangayText.mcrimeind == 1]
dfCrimesText.head()
```

### Get the rows where "brgy_resid_code" isn't null

```python
dfBarangayCode = df[['brgy_resid_code', 'mcrimeind',
       'mtheftind', 'mrapeind', 'minjurind', 'mcarnapind', 'mcattrustlind',
       'mocrimind', 'mocrim', 'mtheftloc', 'mrapeloc', 'minjurloc',
       'mcarnaploc', 'mcattrustlloc', 'mocrimloc']]
dfBarangayCode = dfBarangayCode[dfBarangayCode.brgy_resid_code.notnull()]
dfBarangayCode.head()
```

### Get the rows where mcrimeind = 1 because we only want to see the crimes that happened

```python
dfCrimesCode = dfBarangayCode[dfBarangayCode.mcrimeind == 1]
dfCrimesCode.head()
```

### Combine the dfCrimesCode and dfCrimesText dataframes so we can get the total amount of tracable crimes

```python
dfCrimesFrame = [dfCrimesCode, dfCrimesText]
dfCrimes = pd.concat(dfCrimesFrame, sort = True)
dfCrimes.head()
```

```python
print("Df length: " + str(len(df)))
print("Df Barangays based on Text length: " + str(len(dfBarangayText)))
print("Df Tracable Crimes based on Barangay Text length: " + str(len(dfCrimesText)))
print("Df Barangay based on Code length: " + str(len(dfBarangayCode)))
print("Df Tracable Crimes based on Barangay Code length: " + str(len(dfCrimesCode)))
print("Df Tracable Crimes total: " + str(len(dfCrimes)))
```

```python
index = np.arange(6)
df_sizes = [len(df.index), len(dfBarangayText), len(dfBarangayCode), len(dfCrimes.index), len(dfCrimesText.index), len(dfCrimesCode.index)]
plt.bar(index, df_sizes)
plt.xlabel("Data")
plt.xticks(index, ["Original", "Barangay Txt", "Barangay Code", 
                   "Total Tracable Crimes", "Tracable Crimes by Brgy Txt", 
                   "Tracable Crimes by Brgy Code"], fontsize=10, rotation=30)
plt.ylabel("Data size")
plt.title("Data sizes")
plt.show()
```
